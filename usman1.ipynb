```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
c = pd.read_csv("E:countries.csv")
```

```python
peru=c[c.country == 'Peru']
peru
```

```python
senegal=c[c.country == 'Senegal']
senegal
```

```python
spain=c[c.country == 'Spain']
spain
```

```python
plt.plot(peru.year , peru.population /peru.population.iloc[0] * 100)
plt.plot(senegal.year , senegal.population / senegal.population.iloc[0] * 100)
plt.plot(spain.year , spain.population / spain.population.iloc[0] * 100)

plt.legend(['Peru','Senegal' , 'Spain'])
plt.xlabel('year')
plt.ylabel('Population growth (first year = 100)')
plt.show()
```
